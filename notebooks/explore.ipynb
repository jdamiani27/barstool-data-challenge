```python
import gzip
import json


aa_all_lines = []

with gzip.open("../data/aa.jsonl.gz") as f:
    for line in f:
        aa_all_lines.append(json.loads(line))

len(aa_all_lines)
```

```python
aa_all_lines[0]
```

```python
import zlib

dec = zlib.decompressobj(32 + zlib.MAX_WBITS)
chunk_size = 2**10
i = 0
aa_content_decompressed = b""

while i*chunk_size <= len(aa_content):
    aa_content_decompressed += dec.decompress(aa_content[i*chunk_size:(i+1)*chunk_size])
    i += 1

aa_content_decompressed[:200]
```

```python
len(aa_content_decompressed.strip().split(b"\n"))
```

```python
assert aa_all_lines == [json.loads(l) for l in aa_content_decompressed.strip().split(b"\n")]
```

```python
aa_content_stream = requests.get(aa_url, stream=True)

dec = zlib.decompressobj(32 + zlib.MAX_WBITS)
chunk_size = 2**10
aa_content_stream_decompressed = b""
aa_all_lines_stream = []

for chunk in aa_content_stream.iter_content(chunk_size=chunk_size):
    aa_content_stream_decompressed += dec.decompress(chunk)
    while True:
        index_of_newline = aa_content_stream_decompressed.find(b"\n")

        if index_of_newline == -1:
            break

        aa_all_lines_stream.append(json.loads(aa_content_stream_decompressed[:index_of_newline]))
        aa_content_stream_decompressed = aa_content_stream_decompressed[index_of_newline + 1:]
        
assert aa_all_lines == aa_all_lines_stream
```
